# **Data normalization**
- Uniform the data
- Making sure they have same impact
- Aik machli samandar main aur aik jar main
- Also for computational reasons

```python
kashti.head()
```

```python
ks4 = kashti[["age in days", "fare"]]
ks4.head()
```

- The above data is really in wide range and we need to normalize and hard to compare 
- Normalization change the values to the range of 0-to-10 (now both variables has similar influence on our models)
# **Method of Normalization**
1. Simple feature scaling
   

    - x(new)=x(old)/x(max)


2. Min-Max method
3. Z-score (standard score) -3 -to- +3
4. Log Transformation

```python
# simple feature scaling
ks4['fare'] = ks4['fare']/ks4['fare'].max()
ks4['age in days'] = ks4['age in days']/ks4['age in days'].max()
ks4.head()
```

```python
# Min Max Method
ks4['fare'] = (ks4['fare']-ks4['fare'].min()) / (ks4['fare'].max()-ks4['fare'].max())
ks4.head()
```

```python
ks['fare'] = np.log(ks['fare'])
ks['age in days'] = ks['age in days']/ks['age in days'].max()
ks.head()
```

# Binning 
- Grouping of values into smaller number of values (bins)
-  Convert numeric into categories (jawan, bachay, boorahy) or 1-16, 17-30 etc
-  To have better understanding of groups
   -  low vs mid vs high price

```python
bins = np.linespace(min(kashti['age']), max(kashti['age']),15000)
age_groups = ['bachay', 'jawan', 'boorahy']
kashti['age'] = pd.cut(kashti['age'], bins, labels=age_groups, include_lowest=True)
kashti['age']
```
