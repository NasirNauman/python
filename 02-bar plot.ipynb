```python
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt
# load data set
kashti= sns.load_dataset("titanic")
phool
sns.barplot(x="who", y="alone", hue="sex", data=kashti, color="pink", estimator=median)
plt.title("iris plot")
plt.show()
```

```python
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt
# load data set
kashti= sns.load_dataset("titanic")
phool
sns.barplot(x="who", y="alone", hue="sex", data=kashti, color="red", estimator=median, saturation=2)
plt.title("iris plot")
plt.show()
```
